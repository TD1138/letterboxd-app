```python
import numpy as np
import pandas as pd
```

```python
custom_sorting_dict = {
    'unoptimised':1,
    'remove side bar sliders':2,
    'algo scores to initial load':3,
    'shap scores to initial load':4,
    'stats scatter chart to initial load':5,
    'Actor scatter chart to initial load':6,
    'Move all charts to initial load':7,
    'Add text input to Enter Actor':8,
    'Add text input to Enter Actor 20 actor fix':9,
    'remove some dfs from watchlist tab':10,
    'limit films to show in scatter':11,
    'limit diary to last 100 films':12
    }
custom_sorting_dict = {x+'_initial_data_load': y for x, y in custom_sorting_dict.items()} | {x+'_interaction': y+0.5 for x, y in custom_sorting_dict.items()}
```

```python
timings_df = pd.read_csv('timings_output.csv').fillna(0)
timings_df['run_info'] = timings_df['run_info'] + np.where(timings_df['initial_data_load']>10, '_initial_data_load', '_interaction')
# print('Raw Data:')
# display(timings_df)
initialdataload_df = timings_df[timings_df['run_info'].str.contains('initial_data_load')]
print('Raw Seconds (Initial Dataload):')
display(initialdataload_df.drop(['run_num'], axis=1).groupby('run_info').mean().reset_index().sort_values(by='run_info', key=lambda x: x.map(custom_sorting_dict)))
initialdataload_percent_df = pd.concat([initialdataload_df[['run_num', 'run_info']], initialdataload_df.iloc[:,2:-1].divide(initialdataload_df['total_seconds'], axis=0)], axis=1)
display('Percent of Total Run (Initial Dataload):')
display(initialdataload_percent_df.drop(['run_num'], axis=1).groupby('run_info').mean().reset_index().sort_values(by='run_info', key=lambda x: x.map(custom_sorting_dict)))

interactions_df = timings_df[timings_df['run_info'].str.contains('interaction')]
print('Raw Seconds (Interactions):')
display(interactions_df.drop(['run_num'], axis=1).groupby('run_info').mean().reset_index().sort_values(by='run_info', key=lambda x: x.map(custom_sorting_dict)))
interactions_percent_df = pd.concat([interactions_df[['run_num', 'run_info']], interactions_df.iloc[:,2:-1].divide(interactions_df['total_seconds'], axis=0)], axis=1)
display('Percent of Total Run (Interactions):')
display(interactions_percent_df.drop(['run_num'], axis=1).groupby('run_info').mean().reset_index().sort_values(by='run_info', key=lambda x: x.map(custom_sorting_dict)))
```
